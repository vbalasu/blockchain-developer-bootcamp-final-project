### JSON Schema Validation

```python
from jsonschema import validate, ValidationError
```

```python
schema = {
  "type": "object",
  "properties": {
    "approver_name": {
      "description": "Name of the approver",
      "type": "string",
      "minLength": 2,
      "maxLength": 128
    },
    "approver_email": {
      "description": "Email address of the approver",
      "type": "string",
      "format": "email",
      "pattern": "^\\S+@\\S+\\.\\S+$",
      "format": "email",
      "minLength": 6,
      "maxLength": 127
    }
  },
  "required": [
    "approver_name",
    "approver_email"
  ]
}
```

```python
instance = {"approver_name": "JB", "approver_email": "johnexample.com"}
```

```python
try:
    validate(instance, schema)
except ValidationError as err:
    print(err)
```

```python
%%javascript
var schema = {
  "type": "object",
  "properties": {
    "approver_name": {
      "description": "Name of the approver",
      "type": "string",
      "minLength": 2,
      "maxLength": 128
    },
    "approver_email": {
      "description": "Email address of the approver",
      "type": "string",
      "format": "email",
      "pattern": "^\\S+@\\S+\\.\\S+$",
      "format": "email",
      "minLength": 6,
      "maxLength": 127
    }
  },
  "required": [
    "approver_name",
    "approver_email"
  ]
};
var instance = {"approver_name": "JB", "approver_email": "johnexample.com"};
var Ajv = require("ajv");
var ajv = new Ajv({allErrors: true});
var validate = ajv.compile(schema);
var valid = validate(instance);
if (!valid) console.log(validate.errors);
```
